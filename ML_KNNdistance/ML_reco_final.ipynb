### Ajustement de la moyenne des notes 

```python
# Calculer la moyenne globale des Ã©valuations
C = df_movies1['averageRating'].mean()

# DÃ©finir une constante M (nombre moyen de votes pour qu'un film soit considÃ©rÃ© significatif)
M = df_movies1['numVotes'].mean()

# Ajuster les moyennes en fonction du nombre de votes
df_movies1['adjustedRating'] = (df_movies1['averageRating'] * df_movies1['numVotes'] + C * M) / (df_movies1['numVotes'] + M)
```

# Scaler des colonnes afin de les normaliser

```python

# SÃ©parer les colonnes en groupes de pondÃ©ration
high_weight_columns = list(df_movies1.columns[df_movies1.columns.get_loc('Adventure'):df_movies1.columns.get_loc('Romance') + 1]) 
medium_weight_columns = ['adjustedRating']
standard_weight_columns = ['runtimeMinutes', 'numVotes']

# Appliquer RobustScaler Ã  toutes les colonnes
scaler = RobustScaler()
df_movies1[high_weight_columns + medium_weight_columns + standard_weight_columns] = scaler.fit_transform(df_movies1[high_weight_columns + medium_weight_columns + standard_weight_columns])

# Appliquer les facteurs de pondÃ©ration
high_weight_factor = 15000
medium_weight_factor = 5000
standard_weight_factor = 1

df_movies1[high_weight_columns] = df_movies1[high_weight_columns] * high_weight_factor
df_movies1[medium_weight_columns] = df_movies1[medium_weight_columns] * medium_weight_factor
df_movies1[standard_weight_columns] = df_movies1[standard_weight_columns] * standard_weight_factor


```

### ML pour la reco 

```python
# Train test split

X = df_movies1.drop(columns=['tconst','averageRating', 'titleType', 'title','region', 'language', 'poster_path'])
y = df_movies1['title']
```

```python
# Your code here

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state = 42)
print("The length of the initial dataset is :", len(X))
print("The length of the train dataset is   :", len(X_train))
print("The length of the test dataset is    :", len(X_test))
```

```python
modelNN = NearestNeighbors(n_neighbors=3)
modelNN.fit(X_train, y_train)
```

```python
list_movie = ['Inception']
```

```python
for movie in list_movie:
    # for a movie, pass only same columns as X that our model needs
    neighbors = modelNN.kneighbors(df_movies1.loc[df_movies1['title'] == movie, X.columns])
    print(f"Recommandations for movie {movie} :")
    # find row number (not row name) from the nearest neighbors into dataframe on which the model was fitted
    closest_pok_ind = neighbors[1][0]
    closest_pok = df_movies1['title'].iloc[closest_pok_ind]
    print("Closest Moviess : ", list(closest_pok))
    print("Respectives distances : ", neighbors[0][0])
    print()
```
