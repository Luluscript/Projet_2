```python
import pandas as pd
file_path = r"C:\Users\Lulu\Documents\WCS\Projet_2\tmdb_full.csv"
 # Lire le fichier CSV
df8 = pd.read_csv(file_path, sep=',')
df3 = pd.read_csv("https://datasets.imdbws.com/title.basics.tsv.gz", sep = "\t")
```

## Merge IMDB (title basics) et TMDB

```python
df_merge = pd.merge(df8,
                df3,
                how="inner",
                left_on='imdb_id',
                right_on='tconst',
                sort = True)
```
