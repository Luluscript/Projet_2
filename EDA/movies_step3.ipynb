### DF pour Streamlite et les choix de titre 

```python
df_basics = pd.read_csv("https://datasets.imdbws.com/name.basics.tsv.gz", sep = "\t", usecols=['primaryName','primaryProfession','knownForTitles'], dtype={'primaryName':'str'})
df_basics.head(10)
```

```python
df_merge = pd.merge(df_movies1,
                df_basics,
                how="right",
                left_on='tconst',
                right_on='knownForTitles',
                sort = True)
```

```python
df_clean = df_merge.dropna()
```

```python
df_clean.head(50)
```

```python
# Trier le DataFrame par averageRating (descendant) et numVotes (descendant)
df_sorted2 = df_clean.sort_values(by=['averageRating', 'numVotes'], ascending=[False, False])

# Sélectionner les 1000 premiers films
top_1001 = df_sorted2.head(1000)
```
