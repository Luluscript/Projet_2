### Table name basics

```python
df_basics = pd.read_csv("https://datasets.imdbws.com/name.basics.tsv.gz", sep = "\t", usecols=['primaryName','primaryProfession','knownForTitles'], dtype={'primaryName':'str'})
df_basics.head(10)
```

```python
df_basics.info(memory_usage='deep')
```

```python
# Étape 1: Séparer les chaînes en listes
df_basics['knownForTitles'] = df_basics['knownForTitles'].str.split(',')

# Étape 2: Éclater les listes en lignes distinctes
df_exploded = df_basics.explode('knownForTitles')
```

```python
df_exploded.head()
```

```python
# Spécifiez le chemin du fichier CSV
file_path = 'C:/Users/Lulu/Documents/Git/Projet_2/movies_step3.csv'

# Enregistrer le DataFrame en fichier CSV
df_exploded.to_csv(file_path)

print(f"Le fichier CSV a été enregistré à : {file_path}")
```
