```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc
```

```python
df_ratings = pd.read_csv("https://datasets.imdbws.com/title.ratings.tsv.gz", sep = "\t")

df_imdb = pd.read_csv("https://datasets.imdbws.com/title.basics.tsv.gz", sep = "\t").query("titleType == 'movie'")
df_imdb.head(10)
```

```python
df_imdb.info()
```

```python
df_ratings.info()
```

```python
df_ratings.isna().sum()
```

```python
# MErge des deux tables 

df_merge = pd.merge(df_imdb,
                df_ratings,
                how="right",
                left_on='tconst',
                right_on='tconst',
                sort = True)
```

```python
del df_ratings
gc.collect()
```

```python
del df_imdb
gc.collect()
```

```python
df_merge.head(20)
```

```python
df_merge = df_merge.dropna()
```

```python
df_merge = df_merge.drop(columns= ['endYear'])
```

```python
df_merge.head()
```

```python
df_merge = df_merge[df_merge['isAdult']== 0] 
```

```python
df_merge = df_merge.drop(df_merge[df_merge['genres'] == '\\N'].index)
```

```python
df_merge.head()
```

```python
print(df_merge.dtypes)
```

```python
# Séparation des valeurs de la colonne 'Genres' en listes d'éléments
df_merge['genres'] = df_merge['genres'].str.split(',')

# Création d'une liste de tous les genres uniques
all_genres = set(genre for sublist in df_merge['genres'] for genre in sublist)

# Création des colonnes binaires pour chaque genre unique
for genre in all_genres:
    df_merge[genre] = df_merge['genres'].apply(lambda x: 1 if genre in x else 0)

# Suppression de la colonne originale 'Genres'
df_merge.drop(columns=['genres'], inplace=True)
```

```python
df_merge.head()
```

```python
df_merge.info()
```

```python

# Suppression des lignes où au moins une des colonnes contient 1
df1 = df_merge.drop(df_merge[df_merge['Reality-TV'] == 1].index)
```

```python
del df_merge
gc.collect()
```

```python
df1 = df1.drop(df1[df1['News'] == 1].index)
```

```python
df1 = df1.drop(df1[df1['Music'] == 1].index)
```

```python
df1 = df1.drop(df1[df1['Talk-Show'] == 1].index)
```

```python
df1 = df1.drop(df1[df1['Game-Show'] == 1].index)
```

```python
df1 = df1.drop(df1[df1['Sport'] == 1].index)
```

```python
df1 = df1.drop(columns = ['Reality-TV','News','Music', 'Talk-Show', 'Game-Show', 'Sport'])
```

```python
# Conversion de la colonne 'startYear' en type int
df1['startYear'] = pd.to_numeric(df1['startYear'], errors='coerce')

# Filtrage des années entre 1960 et 2023
df1 = df1[(df1['startYear'] >= 1960) & (df1['startYear'] <= 2023)]
```
